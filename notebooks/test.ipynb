```python
### for importing from same package
import sys
module_path = '../models/'
sys.path.append(module_path)

import pandas as pd

from naver_maps_api import getDirectionsCar, getDirectionsPt, plotNaverMaps
```

```python
def makeSimpleOdDf(start_lat, start_long, goal_lat, goal_long):
  df = pd.DataFrame({
    'startLat': [start_lat],
    'startLong': [start_long],
    'goalLat': [goal_lat],
    'goalLong': [goal_long],
  })
  return df

test = pd.concat([
  makeSimpleOdDf(
    start_lat = 37.560187,
    start_long = 126.936063,
    goal_lat = 37.554938,
    goal_long = 126.972188,
  ), makeSimpleOdDf(
    start_lat = 37.564587,
    start_long = 126.936063,
    goal_lat = 37.554938,
    goal_long = 126.972188,
  )
])
```

```python
getDirectionsCar(test)
```

```python
getDirectionsPt(test)
```

```python
plotNaverMaps(test, markers_latitude_column='startLat', markers_longitude_column='startLong')
```
